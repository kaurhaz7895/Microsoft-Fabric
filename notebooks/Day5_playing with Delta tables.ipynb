```python
%%sql
DESCRIBE HISTORY Public_holidays
```

```python
%%sql
SELECT 
count(*) as Nof_records 
,countryRegionCode
FROM Public_holidays
GROUP BY countryRegionCode
ORDER BY countryRegionCode ASC
```

```python
%%sql
DELETE
FROM Public_holidays
WHERE countryRegionCode = 'SI'
```
